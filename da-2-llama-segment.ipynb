# Import packages for LLaMa

```python
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
from IPython.display import display, Markdown
```

Load LLaMa

```python
model = "/kaggle/input/llama-3/transformers/8b-chat-hf/1"

#instantiate NLP pipeline for huggingface transformers, using default parameters.
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

# import time to measure execution time for both

```python
from time import time
```

define function to generate response from LLaMa, as well as measure the execution time of snippets

```python
def measure_response_time(model, prompt):
    start = time()
    sequences = model(
        prompt,
        do_sample=True,
        top_k=10,
        temperature=0.7,
        num_return_sequences=1,
        eos_token_id=pipeline.tokenizer.eos_token_id,
        max_length=512,
    )
    elapsed_time = time() - start
    return sequences[0]["generated_text"][len(prompt):], elapsed_time
```

Iterate through all questions 5 times, recording the time and average time at the end.

```python
num_iterations = 5
questions = [
    "what is sklearn?",
    "what is linear regression?",
    "perform a linear regression using sklearn given the data [[1,1],[1,2],[2,2],[2,3]]"
]

for question in questions:
    total_elapsed_time = 0
    for i in range(num_iterations):
        response, elapsed_time = measure_response_time(pipeline, question)
        print(f"Iteration {i+1} - Question: {question}")
        print(f"Response: {response}")
        print(f"Time taken: {elapsed_time} seconds")
        total_elapsed_time += elapsed_time
    average_time = total_elapsed_time / num_iterations
    print(f"Average time for question '{question}': {average_time} seconds\n")
```
